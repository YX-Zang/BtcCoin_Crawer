```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
```

```python
url = 'https://graphs2.coinmarketcap.com/currencies/bitcoin/'
res = requests.get(url)
```

```python
btc = res.json()
```

```python
df = pandas.DataFrame(btc)
df = df['price_usd']
date=[]
price=[]
for i in df:
    i[0] = pandas.to_datetime(i[0],unit='ms')
    date.append(i[0])
    price.append(i[1])
    
plt.plot(date,price)
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.title('Price of Bitcoin')
plt.show()
```
